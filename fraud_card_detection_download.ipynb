# Choosing K Value
#### use elbow method to pick a good K value



```python
new_y_test = y_test.iloc[0:100]
new_y_test
```

```python
new_x_test = x_test.iloc[0:100]
new_x_test
```

```python
new_xtrain = x_train.iloc[0:100]
new_xtrain
new_ytrain = y_train.iloc[0:100]

```

```python
# knn = knn.fit(new_xtrain, new_ytrain)
```

```python
# new_pred = knn.predict(new_x_test)
```

```python
error_rate = []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(new_xtrain, new_ytrain)
    pred_i = knn.predict(new_x_test)
    error_rate.append(np.mean(pred_i != new_y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate, color = 'blue', linestyle = '--', marker = 'o', markerfacecolor = 'red', markersize = 10)
plt.title('Error Rate vs K')
plt.xlabel('K')
plt.ylabel("Error Rate")
```
